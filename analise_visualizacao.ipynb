### Análise de Consumo de Energia e Emissões de CO2
#### Tarefas Realizadas
1. **Visualizações Comparativas:**
   - Boxplots para consumo de energia por continente.
   - Gráficos de linha para evolução do consumo e emissões.
   - Gráficos de dispersão com linha de regressão para correlação entre energia fóssil e emissões.
2. **Exploração Regional e Setorial:**
   - Contribuição de setores (indústria, transporte, etc.) nas emissões.
   - Diferenças regionais no impacto do consumo de energia.
3. **Narrativa e Conclusões:**
   - Análise das tendências observadas.
   - Discussão sobre sustentabilidade e transição energética.

```python
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go 
import seaborn as sns
import pandas as pd

df_energy_co2 = pd.read_csv('data_root/dados_energia_simulados_ptbr.csv')
print(df_energy_co2.columns)
```

```python
# Boxplot do consumo de energias renováveis e fósseis por continente
plt.figure(figsize=(10,6))
sns.boxplot(x='Continente', y='Consumo_Energia_Renovável (MWh)', data=df_energy_co2, color="powderblue", showfliers=False)
sns.boxplot(x='Continente', y='Consumo_Energia_Fóssil (MWh)', data=df_energy_co2, color="darkslategray", showfliers=False)
plt.title('Comparação do Consumo de Energias Renováveis e Fósseis por Continente')
plt.ylabel('Consumo de Energia (MWh)')
plt.show()
```

```python
# Gráfico de linha para evolução do consumo de energias e emissões de CO2
plt.figure(figsize=(12,8))
sns.lineplot(data=df_energy_co2, x='Ano', y='Consumo_Energia_Renovável (MWh)', hue='Continente', marker='o')
sns.lineplot(data=df_energy_co2, x='Ano', y='Consumo_Energia_Fóssil (MWh)', hue='Continente', marker='s')
sns.lineplot(data=df_energy_co2, x='Ano', y='Emissões_CO2 (toneladas)', hue='Continente', marker='d')

plt.title('Evolução do Consumo de Energias e Emissões de CO2 ao Longo do Tempo por Continente')
plt.ylabel('Consumo de Energia (MWh) / Emissões de CO2 (toneladas)')
plt.legend(title='Continente')
plt.show()
```

```python
# Gráfico de barras empilhadas das contribuições dos setores
df_grouped = df_energy_co2.groupby(['Continente', 'Setor'])['Emissões_CO2 (toneladas)'].sum().unstack()
df_grouped.plot(kind='bar', stacked=True, figsize=(12,7))
plt.title('Contribuição dos Setores para as Emissões de CO2 por Continente')
plt.ylabel('Emissões de CO2 (toneladas)')
plt.show()
```

```python
# Gráfico interativo de dispersão
fig = px.scatter(df_energy_co2, 
                 x='Consumo_Energia_Renovável (MWh)', 
                 y='Emissões_CO2 (toneladas)', 
                 color='Continente', 
                 animation_frame='Ano', 
                 title='Impacto do Consumo de Energias no CO2 por Continente ao Longo do Tempo',
                 labels={'Consumo_Energia_Renovável (MWh)': 'Consumo de Energia Renovável (MWh)',
                         'Emissões_CO2 (toneladas)': 'Emissões de CO2 (toneladas)'})

# Adicionando outra série para consumo de energia fóssil
fig.add_trace(px.scatter(df_energy_co2, 
                         x='Consumo_Energia_Fóssil (MWh)', 
                         y='Emissões_CO2 (toneladas)', 
                         color='Continente', 
                         animation_frame='Ano', 
                         labels={'Consumo_Energia_Fóssil (MWh)': 'Consumo de Energia Fóssil (MWh)'}).data[0])

# Atualizando o layout
fig.update_layout(title='Impacto do Consumo Energético nas Emissões de CO2 por Continente ao Longo do Tempo',
                  xaxis_title='Consumo de Energia (MWh)',
                  yaxis_title='Emissões de CO2 (toneladas)',
                  legend_title='Continente')

fig.show()
```

#### Análise e Relação entre Consumo de Energia e Emissões de CO2:
Consumo e Emissões: 
Quanto mais energia fóssil um país consome, maior é a emissão de CO2. A queima de combustíveis fósseis para gerar energia é uma grande responsável pelo aumento dos gases de efeito estufa. Por isso, investir em energias renováveis, como solar e eólica, é super importante! 

Diferenças Regionais: 
Regiões que apostam em energias renováveis geralmente têm uma pegada de carbono menor. Já aquelas que ainda dependem de combustíveis fósseis enfrentam níveis mais altos de emissões. Isso mostra a importância de cada lugar encontrar seu caminho rumo à sustentabilidade. 

Setores de Consumo: 
O setor em que a energia é usada também faz diferença. O transporte, que depende bastante de combustíveis fósseis, tem altas emissões, enquanto setores que usam mais energia renovável tendem a ser mais sustentáveis.

Conclusão:
Essas visualizações nos ajudam a perceber como o consumo energético e as emissões de CO2 se relacionam, destacando as diferenças entre regiões e setores. Com isso, podemos pensar em soluções para um futuro mais sustentável!
